## Tuples


#### A tuple is an ordered, immutable collection of items in Python. Immutability means you cannot change, add, or remove elements after creation.

```python
t = (10, 20, 30)
```

```python
t = (10,)     # tuple
x = (10)      # NOT a tuple (just an integer)
```

```python
t = ("Anil", 25, 55.6, True) # with mixed data types
print(t)
```

```python
#indexing
t = ("apple", "banana", "cherry")
print(t[0])
print(t[-1])
```

```python
# Immutability
t = (10, 20, 30)
t[1] = 50
```

```python

t2 = t + (40,)  # create new tuple
print(t2)       # (10, 20, 30, 40)
```

```python
## 2 methods in tuples
# 1. count
t = (1, 2, 2, 3, 2)
print(t.count(2))

#index()
t = ("a", "b", "c", "b")
print(t.index("b"))
```

## Sets

#### A set is an unordered, mutable, and unique-element collection in Python.

```python
s = {1, 2, 3}
s
```

```python
## sets automatically delete duplicates
s = {1, 2, 2, 3}
print(s)
```

```python
# set from list
s = set([1, 2, 3])
print(s)
```

```python
# Mixed data Types
s = {10, "apple", (1, 2)}
print(s)
```

```python
# adding new element
s = {1, 2}
s.add(3)
print(s)
```

```python
# add multiple elements
s = {1, 2}
s.update([3, 4, 5])
print(s)
```

```python
## to remove an element
s = {1, 2, 3}
s.remove(2)
print(s)
```

```python
# no error if element is not present
s = {1, 2, 3}
s.discard(10)
print(s)
```

```python
# pop removes random element
s = {1, 2, 3}
print(s.pop())
print(s)
```

```python
s.clear()
print(s)
```

## Dictionaries


#### A dictionary is a mutable, unordered collection of key–value pairs.
--> Keys must be unique

--> Keys must be immutable (string, number, tuple)

--> Values can be any type

```python
student = {
    "name": "Anil",
    "age": 25,
    "marks": 95
}
print(student)
```

```python
# creating a dictionary
d = {"a": 1, "b": 2}
print(d)

# Using Dict
d2 = dict(name="Anil", age=22)
print(d2)
```

```python
# mixed data types

d = {"name": "Anil", "age": 25, "skills": ["Python", "ML"]}
print(d)
```

```python
# Accessing Elements
# 1. using key
student = {"name": "Anil", "age": 22}
print(student["name"])

# 2. using get()

print(student.get("name"))
```

```python
# Adding new element
d = {"a": 1}
d["b"] = 2
print(d)

# updating
d["a"] = 10
print(d)
```

```python
# remove specific element
d = {"a": 1, "b": 2}
d.pop("a")
print(d)
```

```python
# removes last inserted key
d = {"a": 1, "b": 2}
d.popitem()
print(d)
```

```python
# delete a key
d = {"a": 1, "b": 2}
del d["b"]
print(d)

#empty the dictionary
d.clear()
print(d)
```

```python
# loop keys
d = {"a": 1, "b": 2,"c":3}
for k in d:
    print(k)
```

```python
# loop values
d = {"a": 1, "b": 2,"c":3}
for v in d.values():
    print(v)
```

```python
# key value pairs
for k, v in d.items():
    print(k, v)
```

```python
# nested dictionary
student = {
    "name": "Anil",
    "marks": {
        "math": 90,
        "python": 95
    }
}
print(student["marks"]["python"])
```

### Dictionary comprehension

```python
# A concise way to create dictionaries.
squares = {x: x*x for x in range(1, 6)}
print(squares)
```

### list comprehension


```python
# without comprehension
nums = [1, 2, 3, 4, 5, 6]
evens = []

for n in nums:
    if n % 2 == 0:
        evens.append(n)

print(evens)
```

```python
# with comprehension
nums = [1, 2, 3, 4, 5, 6]
evens = [n for n in nums if n % 2 == 0]
print(evens)   # [2, 4, 6]
```

### shallow copy
#### A shallow copy creates a new outer object,but the inner elements are not copied — only references are copied.

Meaning:
Changes to inner elements affect both copies.
Changes to outer list (append/remove) affect only one.

```python
a = [[1, 2], 3]
b = a.copy()

b[0][1] = 99

print(a)
print(b)
```

### Deep Copy
A deep copy creates a new outer object and
recursively copies all inner elements.

Meaning:

Completely independent copy

Changing inner elements does NOT affect the original

```python
import copy

a = [[1, 2], 3]
b = copy.deepcopy(a)

b[0][1] = 99

print(a)
print(b)
```

## Functions

#### A function is a reusable block of code that performs a specific task.

```python
# difining the function
def greet():
    print("Hello!")

# calling the function
greet()
```

```python
# function with parameters
def add(a, b):
    return a + b

print(add(3, 5))
```

Parameters vs Arguments

Parameter → variable in function definition

Argument → value you pass when calling

A function can return:

single value

multiple values (as tuple)

```python
def multiply(a, b):
    return a * b

result = multiply(4, 5)
print(result)
```

```python
#multiple returns
def calc(a, b):
    return a+b, a-b

x, y = calc(10, 5)
print(x, y)
```
